### Functions:

Function to calculate the separation betweem two approximately Gaussian distributions (see equation at the bottom of the script).

__Note__: You need to fill in this function! 

```python
def calc_separation(x, y):
    
    mu_x = np.mean(x)
    mu_y = np.mean(y)
    std_x = np.std(x)
    std_y = np.std(y)
    
    z = np.abs(mu_x - mu_y) / np.sqrt(std_x**2 + std_y**2)
    
    return z
```

## Define parameters:


Number of species, their means and widths, correlations and the number of observations of each species:

```python
# Number of 'species': signal / background
n_spec = 2       

# Species A, mean and width for the two dimensions/parameters
mean_A  = [15.0, 50.0] 
width_A = [ 2.0,  6.0] 

# Species B, mean and width for the two dimensions/parameters
mean_B  = [12.0, 55.0] 
width_B = [ 3.0,  6.0] 

# Coefficient of correlation
corr_A = 0.8
corr_B = 0.9

# Amount of data you want to create
n_data = 2000         
```

## Generate data:

For each "species", produce a number of $(x_0,x_1)$ points which are (linearly) correlated:

```python
# The desired covariance matrix.
V_A = np.array([[width_A[0]**2, width_A[0]*width_A[1]*corr_A],
                [width_A[0]*width_A[1]*corr_A, width_A[1]**2]])
V_B = np.array([[width_B[0]**2, width_B[0]*width_B[1]*corr_B],
                [width_B[0]*width_B[1]*corr_B, width_B[1]**2]])

# Generate the random samples.
spec_A = np.random.multivariate_normal(mean_A, V_A, size=n_data)
spec_B = np.random.multivariate_normal(mean_B, V_B, size=n_data)
```

## Fisher Discriminant calculation:

We want to find $\vec{w}$ defined by:

$$\vec{w} = \left(\Sigma_A + \Sigma_B\right)^{-1} \left(\vec{\mu}_A - \vec{\mu}_B\right)$$

which we use to project our data into the best separating plane (line in this case) given by:

$$ \mathcal{F} = w_0 + \vec{w} \cdot \vec{x} $$

We start by finding the means and covariance of the individuel species: (__fill in yourself!__)

```python
mu_A_x0 = np.mean(spec_A[:,0])
mu_A_x1 = np.mean(spec_A[:,1])
mu_A = np.array([mu_A_x0, mu_A_x1])

mu_B_x0 = np.mean(spec_B[:,0])
mu_B_x1 = np.mean(spec_B[:,1])  
mu_B = np.array([mu_B_x0, mu_B_x1])

mu_A
```

```python
cov_A = np.cov(spec_A[:,0],spec_A[:,1])
cov_B = np.cov(spec_B[:,0],spec_B[:,1])

cov_A
```

```python
cov_sum = cov_A + cov_B
cov_sum
```

where `cov_sum` is the sum of the all of the species' covariance matrices. We invert this using scipy's `inv` function.  __Note__: fill in yourself!

```python


# Inverts cov_sum
cov_sum_inv = inv(cov_sum)
cov_sum_inv
```

We calculate the fisher weights, $\vec{w}$. __Note__: fill in yourself:

```python
wf = cov_sum_inv @ (mu_A - mu_B)
wf
```

We calculate the fisher discriminant, $\mathcal{F}$. __Note__: fill in yourself:

```python
fisher_data_A = spec_A @ wf  # fill in yourself
fisher_data_B = spec_B@ wf  # fill in yourself

```

and plot it:

```python
fig_fisher, ax_fisher = plt.subplots(figsize=(12, 8))
ax_fisher.hist(fisher_data_A, 100, range=(-30, 20), histtype='step', color='Red', label='Species A')
ax_fisher.hist(fisher_data_B, 100, range=(-30, 20), histtype='step', color='Blue', label='Species B')
ax_fisher.set(xlabel='Fisher-discriminant', ylabel='Frequency / 0.5')
ax_fisher.legend(fontsize=20)

# ax_fisher.text(-21, 60, fr'$\Delta_{{fisher}} = {calc_separation(fisher_data_A, fisher_data_B):.3f}$', fontsize=16)

fig_fisher.tight_layout()

if save_plots:
    fig_fisher.savefig('FisherOutput.pdf', dpi=600)
```

```python
fig_corr, ax_corr = plt.subplots(figsize=(14, 8))
def wf_line(x):
    return (-wf[0]/wf[1]) * x + (mu_A @ wf + mu_B @ wf) / (2 * wf[1])

ax_corr.scatter(spec_A[:, 0], spec_A[:, 1], color='Red',  s=10, label='Species A')
ax_corr.scatter(spec_B[:, 0], spec_B[:, 1], color='Blue', s=10, label='Species B')
ax_corr.set(xlabel='Parameter x0', ylabel='Parameter x1', title='Correlation')

x_vals = np.array(ax_corr.get_xlim())
ax_corr.plot(x_vals, wf_line(x_vals), '--', color='Black', label='Fisher discriminant line')

midpoint = (mu_A + mu_B) / 2
scale = 5  # Adjust this to make arrow visible

ax_corr.arrow(midpoint[0], midpoint[1], 
              scale*wf[0], scale*wf[1],
              head_width=0.5, head_length=0.3, 
              fc='green', ec='green', linewidth=2,
              label='Weight vector w')


ax_corr.set_aspect('equal')
ax_corr.legend();
fig_corr.tight_layout()

if save_plots :
   fig_corr.savefig('InputVars_2D.pdf', dpi=600)
```

```python
# Question 7: Variable Importance Analysis

# Step 1: Calculate the difference in means for each variable
delta_mu = np.abs(mu_A - mu_B)
print("Difference in means:")
print(f"  _x0 = {delta_mu[0]:.3f}")
print(f"  _x1 = {delta_mu[1]:.3f}")

# Step 2: Calculate normalized importance metric
# If variables were equally important, weights should be inversely proportional to delta_mu
# So: w_normalized = w * delta_mu should be similar for both variables
importance_metric = wf * delta_mu
print("\nImportance metric (w * ):")
print(f"  x0: {importance_metric[0]:.3f}")
print(f"  x1: {importance_metric[1]:.3f}")

# Step 3: Compare relative importance
relative_importance = importance_metric / np.max(np.abs(importance_metric))
print("\nRelative importance:")
print(f"  x0: {relative_importance[0]:.3f}")
print(f"  x1: {relative_importance[1]:.3f}")

# Step 4: Test by excluding variables one at a time
print("\n" + "="*50)
print("Testing separation with each variable excluded:")
print("="*50)

# Full separation (already calculated)
print(f"Full Fisher (both variables): {z_fish:.3f}")

# Only x0 (exclude x1) - Fisher reduces to just the x0 projection
z_x0_only = calc_separation(spec_A[:, 0], spec_B[:, 0])
print(f"Only x0 (exclude x1):         {z_x0_only:.3f}")
print(f"  Loss: {z_fish - z_x0_only:.3f} ({100*(z_fish - z_x0_only)/z_fish:.1f}%)")

# Only x1 (exclude x0)
z_x1_only = calc_separation(spec_A[:, 1], spec_B[:, 1])
print(f"Only x1 (exclude x0):         {z_x1_only:.3f}")
print(f"  Loss: {z_fish - z_x1_only:.3f} ({100*(z_fish - z_x1_only)/z_fish:.1f}%)")

# Step 5: Conclusion
print("\n" + "="*50)
print("Conclusion:")
print("="*50)
if np.abs(importance_metric[0]) > np.abs(importance_metric[1]):
    print("x0 is MORE important than x1")
    print(f"  (importance ratio: {np.abs(importance_metric[0]/importance_metric[1]):.2f}:1)")
else:
    print("x1 is MORE important than x0")
    print(f"  (importance ratio: {np.abs(importance_metric[1]/importance_metric[0]):.2f}:1)")

print("\nThis makes sense because:")
print("- A larger weight * delta_mu means the variable contributes more to separation")
print("- Excluding the more important variable causes a larger drop in separation power")
```
